```python
print("Digite seus n√∫meros: ")

num1 = int(input("Digite o primeiro n√∫mero: "))
num2 = int(input("Digite o segundo n√∫mero: "))

def num_aleatorio(): # Tentativa de definir a fun√ß√£o
  # Use global keyword to access the global num1 and num2
  global num1
  global num2

  import random # Importa o m√≥dulo random
  historico_numeros = [] # Lista para armazenar os n√∫meros aleat√≥rios gerados

  # Ensure num1 is the smaller and num2 is the larger value immediately
  if num1 > num2:
     num1, num2 = num2, num1

  print("Selecione o que deseja:")
  print("1. Gerar n√∫mero aleat√≥rio")
  print("2. Mostrar hist√≥rico")
  print("3. Modo Bingo")
  print("4. Sair")

  while True:
      escolha = input("Digite o n√∫mero da op√ß√£o desejada: ")
      if escolha == "1":

             numero_aleatorio = random.randint(num1, num2) # Escolhe um n√∫mero aleat√≥rio entre num1 e num2 (inclusive ambos), Se voc√™ quiser que os n√∫meros dos extremos n√£o sejam inclu√≠dos na escolha, use random.randint(num1 + 1, num2 - 1)
             print(f"O n√∫mero aleat√≥rio escolhido entre {num1} e {num2} √©: {numero_aleatorio}")

          # Add the random number to the history
             historico_numeros.append(numero_aleatorio)

      elif escolha == "2":
          if not historico_numeros:
              print("Hist√≥rico de n√∫meros aleat√≥rios vazio.")
          else:
              print("Hist√≥rico de n√∫meros aleat√≥rios:")
              for numero in historico_numeros:
                  print(numero)

      elif escolha == "3":
        # 2. Criar a cartela de bingo (lista de n√∫meros dispon√≠veis)
        # Usamos range para gerar os n√∫meros e list() para convert√™-los em uma lista
        cartela_bingo = list(range(num1, num2 + 1)) # +1 para incluir o num_fim
        print(f"\nCartela de bingo criada com n√∫meros de {num1} a {num2}.")
        print(f"N√∫meros dispon√≠veis: {cartela_bingo}")
        # Loop principal do bingo
        while cartela_bingo: # O loop continua enquanto houver n√∫meros na cartela
            input("\nPressione Enter para sortear um n√∫mero...") # Pausa para o usu√°rio

            # 3. Sortear um n√∫mero aleat√≥rio da lista de dispon√≠veis
            numero_sorteado = random.choice(cartela_bingo)

            print(f"üéâ N√∫mero sorteado: {numero_sorteado} üéâ")

            # 4. Remover o n√∫mero sorteado da lista para que n√£o seja sorteado novamente
            cartela_bingo.remove(numero_sorteado)

            print(f"N√∫meros restantes: {cartela_bingo}")

            # Verificar se a cartela est√° vazia
            if not cartela_bingo:
                print("\nüèÜ Todos os n√∫meros foram sorteados! Fim do bingo! üèÜ")
                break # Sai do loop quando a cartela estiver vazia

      elif escolha == "4":
          print("Saindo do programa.")
          break
      else:
          print("Op√ß√£o inv√°lida. Por favor, escolha uma op√ß√£o v√°lida.")

num_aleatorio()
```

```python
import random #Gemini

print("Digite seus n√∫meros: ")

# Pede os n√∫meros uma √∫nica vez, antes de chamar a fun√ß√£o
num1 = int(input("Digite o primeiro n√∫mero: "))
num2 = int(input("Digite o segundo n√∫mero: "))

# Garante que num1 seja o menor e num2 o maior valor imediatamente
if num1 > num2:
    num1, num2 = num2, num1

# Definimos as listas de hist√≥rico e a cartela de bingo fora do loop principal
# e da fun√ß√£o, para que persistam durante toda a execu√ß√£o do programa.
historico_numeros_aleatorios = [] # Para n√∫meros aleat√≥rios (op√ß√£o 1)
historico_numeros_bingo = [] # Para n√∫meros sorteados no bingo (op√ß√£o 3)
cartela_bingo_atual = [] # A cartela que ser√° usada no modo bingo

def menu_principal():
    print("\n--- MENU PRINCIPAL ---")
    print("1. Gerar n√∫mero aleat√≥rio")
    print("2. Mostrar hist√≥rico de n√∫meros aleat√≥rios")
    print("3. Iniciar Modo Bingo")
    print("4. Mostrar hist√≥rico de n√∫meros do Bingo")
    print("5. Sair")
    print("----------------------")

def gerenciar_numeros_aleatorios(min_val, max_val, historico):
    """Gera um n√∫mero aleat√≥rio e o adiciona ao hist√≥rico."""
    numero_aleatorio = random.randint(min_val, max_val)
    print(f"O n√∫mero aleat√≥rio escolhido entre {min_val} e {max_val} √©: {numero_aleatorio}")
    historico.append(numero_aleatorio)

def mostrar_historico(historico, titulo):
    """Exibe o hist√≥rico de n√∫meros."""
    if not historico:
        print(f"Hist√≥rico de {titulo} vazio.")
    else:
        print(f"Hist√≥rico de {titulo}:")
        for numero in historico:
            print(numero)

def modo_bingo(min_val, max_val, historico_bingo, cartela_ativa):
    """Inicia e gerencia o modo bingo."""
    # Reinicia a cartela para cada novo jogo de bingo
    cartela_ativa.clear() # Limpa a cartela anterior, se houver
    cartela_ativa.extend(list(range(min_val, max_val + 1)))

    print(f"\n--- MODO BINGO ---")
    print(f"Cartela de bingo criada com n√∫meros de {min_val} a {max_val}.")
    print(f"N√∫meros dispon√≠veis: {cartela_ativa}")

    while cartela_ativa: # O loop continua enquanto houver n√∫meros na cartela
        input("\nPressione Enter para sortear um n√∫mero... (ou digite 'sair' para voltar ao menu)")

        # Permite sair do modo bingo a qualquer momento
        comando = input().strip().lower()
        if comando == 'sair':
            print("Saindo do Modo Bingo e voltando ao menu principal.")
            break

        numero_sorteado = random.choice(cartela_ativa)
        print(f"üéâ N√∫mero sorteado: {numero_sorteado} üéâ")

        cartela_ativa.remove(numero_sorteado)
        historico_bingo.append(numero_sorteado) # Adiciona ao hist√≥rico do bingo imediatamente

        print(f"N√∫meros restantes: {cartela_ativa}")

        if not cartela_ativa:
            print("\nüèÜ Todos os n√∫meros foram sorteados! Fim do bingo! üèÜ")
            break # Sai do loop do bingo

# --- Loop Principal do Programa ---
while True:
    menu_principal()
    escolha = input("Digite o n√∫mero da op√ß√£o desejada: ")

    if escolha == "1":
        gerenciar_numeros_aleatorios(num1, num2, historico_numeros_aleatorios)
    elif escolha == "2":
        mostrar_historico(historico_numeros_aleatorios, "n√∫meros aleat√≥rios")
    elif escolha == "3":
        # Chama a fun√ß√£o modo_bingo
        modo_bingo(num1, num2, historico_numeros_bingo, cartela_bingo_atual)
    elif escolha == "4":
        mostrar_historico(historico_numeros_bingo, "n√∫meros do Bingo")
    elif escolha == "5":
        print("Saindo do programa. At√© mais!")
        break
    else:
        print("Op√ß√£o inv√°lida. Por favor, escolha uma op√ß√£o v√°lida (1 a 5).")
```
